## Object-Oriented Programming

```python
# keep this in a safe place.
apikey_token='e11115c7-7775-4694-a113-6f2910'
```

```python
# Get team-level statistics for the 2018 MLB season

class Baseball:
    apikey_token=apikey_token # defined outside of the class for security reasons.

    def __init__(self, team):
        self.team=team
        self.url='https://api.mysportsfeeds.com/v2.0/pull/mlb/2018-regular/team_stats_totals.json'
    
    # get request
    def get_request(self):
        try:
            response = requests.get(            
                url=self.url,
                params={
                    "team": self.team,    
                },
                headers={
                    "Authorization": "Basic " + base64.b64encode('{}:{}'.format(apikey_token,'MYSPORTSFEEDS').encode('utf-8')).decode('ascii')
                }
            )
            return response.json()
        except requests.exceptions.RequestException:
            print('HTTP Request failed')
    
    # team's full name
    def full_name(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['team']['name']
    
    # name of city
    def city(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['team']['city']
    
    # name of ballpark
    def ballpark(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['team']['homeVenue']['name']
    
    # team batting average
    def batting_avg(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['stats']['batting']['battingAvg']
    
    # team on-base percentage
    def on_base(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['stats']['batting']['batterOnBasePct']
    
    # team earned-run average
    def earned_run(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['stats']['pitching']['earnedRunAvg']
    
    # team winning percentage for regular season
    def win_pct(self):
        response=self.get_request()
        return response['teamStatsTotals'][0]['stats']['standings']['winPct']
```
